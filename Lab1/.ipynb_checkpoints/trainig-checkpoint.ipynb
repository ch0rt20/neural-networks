# Глава 3

## Один вход

```python
weight = 0.1
def neural_network(input, weight):
    prediction = input * weight
    return prediction
```

```python
number_of_toes = [8.5, 9.5, 10, 9]
input = number_of_toes[0]
pred = neural_network(input, weight)
print(pred)
```

## Несколько входов

```python
def w_sum(a, b):
    assert (len(a) == len(b))
    output = 0
    for i in range (len(a)):
        output += (a[i]*b[i])
    return output    
```

```python
weights = [0.1, 0.2, 0]
def neural_network(input, weights):
    pred = w_sum(input, weights)
    return pred
```

```python
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [8.5, 9.5, 9.9, 9.0]
input = [toes[0], wlrec[0], nfans[0]]
pred = neural_network(input, weights)
print(pred)
```

NUMPY

```python
import numpy as np

weights = np.array([0.1, 0.2, 0])
def neural_network(input, weights):
    pred = input.dot(weights)
    return pred
```

```python
import numpy as np

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [8.5, 9.5, 9.9, 9.0]
input = np.array([toes[0], wlrec[0], nfans[0]])
pred = neural_network(input, weights)
print(pred)
```

# Глава 4 

```python
knob_weight = 0.5
input = 0.5
goal_pred = 0.8
pred = 0
step_amount = 0.001
k = 0

while pred < goal_pred:
    pred = input * knob_weight
    error = (pred - goal_pred) ** 2
    
    print ('Error' + str(error) + 'Prediction' + str(pred))
    
    up_pred = input * (knob_weight + step_amount)
    up_err = (goal_pred - up_pred) ** 2
    
    down_pred = input * (knob_weight - step_amount)
    down_err = (goal_pred - down_pred) ** 2
    
    if (down_err < up_err):
        knob_weight =  knob_weight - step_amount
    if (down_err > up_err):
        knob_weight =  knob_weight + step_amount     
    
```
